### calculer the RMSD between two model

```python
def calc_rmsd(reference, reference2):
    total=[]
    for i in range(0,len(reference)):
        liste1=reference[i].split(',')
        x1=float(liste1[1])
        y1=float(liste1[2])
        z1=float(liste1[3])
        liste2=reference2[i].split(',')
        x2=float(liste2[1])
        y2=float(liste2[2])
        z2=float(liste2[3])
        total.append((x1-x2)**2 + (y1-y2)**2 + (z1-z2)**2)
    N=len(reference)
    x=math.sqrt(1/N * sum(total))
    return x
```

```python
def matrix_rmsd(coor_dico):
    MI_dict={}
    for key1,values1 in coor_dico.items():
        MI_list=[]
        for key2,values2 in coor_dico.items():
            x=calc_rmsd(values1, values2)
            MI_list.append(x)
        MI_dict[key1]=MI_list
    dataframe_MI=pd.DataFrame(MI_dict)
    return dataframe_MI
```

```python
dict_all_rmsd={}
for i in files:
    dict_all_rmsd[i]=matrix_rmsd(dict_all_coor[i])
```

```python
dict_len={}
for i in dict_groupe:
    len_cluster=[]
    for j in dict_groupe[i]:
        len_cluster.append(len(dict_groupe[i][j]))
    dict_len[i]=max(len_cluster)
        
```

```python
dict_sum_score={}
for i in dict_groupe:
    dict_sum_score[i]={}
    for j in dict_groupe[i]:
        liste_value=dict_groupe[i][j]
        min_value=min(liste_value)
        mean_value=mean(liste_value)
        score1=abs(min_value-mean_value)/abs(min_value)
        score1=math.sqrt(score1)
        score2=abs(dict_len[i]-len(dict_groupe[i][j]))/dict_len[i]
        score2=math.sqrt(score2)
        dict_sum_score[i][j]=score1+score2
```
